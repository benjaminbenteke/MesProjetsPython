```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1) Linear Regression with sklearn

```python
np.random.seed(0)
```

```python
m=100 # 100 exemples
X=np.linspace(0,10,m).reshape(m,1) # m lignes et une seule colonne
y=X+np.random.randn(m,1)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X,y)
model.score(X,y) # Coefficient de determination R carre de la methode de moindre carree
```

```python
y_pred=model.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,c='r',LW=3)
```

# Model de regression non lineaire Support vectors Regressions (SVR)

```python
from sklearn.svm import SVR
```

```python
modelS=SVR(C=100)
modelS.fit(X,y)
modelS.score(X,y) # Coefficient de determination R carre de la methode de moindre carree
```

```python
y_predS=modelS.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,y_predS,c='r',LW=3)
```
