## Preparation de données de Iris de Fisher

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
X=pd.DataFrame(iris.data)
y=pd.DataFrame(iris.target)
t=iris.target_names
z=iris.feature_names
print(t)
print(z)
```

```python
X.columns=['sepal length','sepal width','petal length','petal width']
```

```python
X.head(2)
```

```python
X['classe']=y
X.head(2)
```

```python
X['classe']=(X['classe']).replace([0,1,2],['setosa','versicolor','virginica'])
```

```python
X.head(2)
```

### Exportation du dataSet

```python
X.to_csv('Iris_data.csv',index=False)
```

# Data Set du Titanic

### Preparation

```python
titanic=sns.load_dataset('titanic')
titanic.to_csv('titanic.csv',index=False)
titanic.head()
```

```python
titanic.drop(['alone','alive','who','embark_town','adult_male','class'],axis=1,inplace=True)
titanic.dropna(axis=0,inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

## Remarks 

### Plot Categorielle: on a des fonctions telles que: catplot, boxplot,...

### Catplot

```python
sns.catplot(x='pclass',y='age',data=titanic,hue='sex')
```

### Boxplot

```python
sns.boxplot(x='pclass',y='age',data=titanic,hue='sex')
```

### Interpretation du boxplot

#### 1) Dans le boxplot, Q1 et Q3 veut dire la majorité.

## Distribution plot (histogramme)
pairplot,distplot,jointplot,...

```python
sns.distplot(titanic['fare'])
```

### jointplot

```python
sns.jointplot('age','fare',data=titanic,kind='kde') # kde: kernel density
```

```python
sns.jointplot('age','fare',data=titanic,kind='hex') #hex:hexagonal
```

## La fonction heatmap
Tres utile pour visualiser des matrice des correlations.

```python
correlation=titanic.corr()
correlation
```

```python
sns.heatmap(correlation)
```
