# Aggregrate Statistics (Groupby)

```python
## On fait groupby pour avoir le resultat par catégorie (region, sexe, pays et autres). par exemple on peut faire la moyenne (std) 
## de chaque region ou de chaque pays.
```

```python
df=pd.read_excel('don.xlsx')
df
```

```python
df.groupby(['pays']).mean()
```

```python
df.groupby('sexe').mean()
```

```python
df.groupby('sexe').std()
```

```python
df.groupby(['pays']).max()['age']
```

```python
df.groupby(['pays','nom']).max()['note']
```

```python
## pour savoir les occurences de chaque pays
```

```python
df.groupby('pays').count()['nom']
```
